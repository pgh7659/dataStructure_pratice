### Binary Search Tree

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class NodeMgmt:
    def __init__(self, head):
        self.head = head
    
    def insert(self, value):
        self.current_node = self.head
        while True :
            print(self.current_node)
            if self.current_node.value > value:
                if self.current_node.left != None:
                    self.current_node = self.current_node.left
                else:
                    self.current_node.left = Node(value)
                    break
            else:
                if self.current_node.right != None:
                    self.current_node = self.current_node.right
                else:
                    self.current_node.right = Node(value)
                    break
                
            
    def search(self, value):
        self.current_node = self.head
        while self.current_node :
            if self.current_node.value == value:
                return self.current_node
            elif self.current_node.value > value:
                self.current_node = self.current_node.left
            else:
                self.current_node = self.current_node.right
        
        return False
    
    def delete(self, value):
        searched = False
        self.current_node = self.head
        self.parent = self.head

        while self.current_node:
            if self.current_node.value == value:
                searched = True
                break
            elif self.current_node.value <= value:
                self.parent = self.current_node
                self.current_node = self.current_node.right
            else:
                self.parent = self.current_node
                self.current_node = self.current_node.left

        if searched:
            #삭제할 node가 leaf node인 경우
            if self.current_node.left == None and self.current_node.right == None:
                if self.parent.value > value:
                    self.parent.left = None
                else:
                    self.parent.right = None
                del self.current_node

            #삭제할 node가 1개의 left child node를 갖는 경우
            elif self.current_node.left != None and self.current_node.right == None:
                self.current_node = self.current_node.left
                if self.parent.value > value:
                    self.parent.left = self.current_node
                else:
                    self.parent.right = self.current_node

            #삭제할 node가 1개의 right child node를 갖는 경우
            elif self.current_node.left == None and self.current_node.right != None:
                self.current_node = self.current_node.right
                if self.parent.value > value:
                    self.parent.left = self.current_node
                else:
                    self.parent.right = self.current_node

            #삭제할 node가 left / right child node 모두를 갖는 경우
            else:
                self.change_node = self.current_node.left
                self.change_parent = self.current_node
                while self.change_node.right:
                    self.change_parent = self.change_node
                    self.change_node = self.change_node.right

                if self.change_node.left:
                    self.change_parent.left = self.change_node.left
                else:
                    self.change_parent.left = None

                self.change_node.left = self.current_node.left
                self.change_node.right = self.current_node.right
                self.current_node = self.change_node
                
                if self.parent.value > value:
                    self.parent.left = self.current_node
                else:
                    self.parent.right = self.current_node
```

```python
head = Node(1)
BST = NodeMgmt(head)
BST.insert(3)
BST.insert(4)
BST.insert(9)
BST.insert(6)
BST.insert(2)
BST.insert(5)
BST.insert(13)
BST.insert(35)
BST.insert(31)
```

```python
BST.delete(13)
```

#### 이진 탐색 트리 삭제

##### 1. LEAF Node인 경우
- 삭제 후 ParentNode의 ChildNode None으로 변경

##### 2. Child Node가 1개인 경우
- 삭제 후 PaentNode에 ChildNode를 이어붙인다

##### 3. Child Node가 2개인 경우
- "작은값(삭제할 Node의 Left Child) 중 최대값" OR "큰값(삭제할 Node의 Right Child)중 최소값" 으로 변경!

### python 전체 코드 테스트
  - random 라이브러리 활용

```python
import random

# 0 - 999 숫자 중에서 임의로 100개를 추출하여, 이진탐색트리 입력,검색,삭제 테스트 진행
bst_numbers = set()
while len(bst_numbers) != 100:
    bst_numbers.add(random.randint(0, 999))

head = Node(500)
binary_tree = NodeMgmt(head)

# 입력 테스트
for num in bst_numbers:
    binary_tree.insert(num)

# 검색 테스트    
for num in bst_numbers:
    if binary_tree.search(num) == False:
        print('Error!!!!', num)

# 입력한 100개의 숫자중 10개 선별하여 삭제 테스트
del_numbers = set()
bst_numbers_list = list(bst_numbers)
while len(del_numbers) != 10:
    del_numbers.add(bst_numbers_list[random.randint(0, 99)])

#for num in bst_numbers:
#    binary_tree.delete(num)

curr = binary_tree.search(500)
while True:
    print(curr.value)
    if curr.right != None:
        curr_right_value = curr.right.value
        curr = binary_tree.search(curr_right_value)
    else:
        break    
```

### 이진탐색트리의 시간복잡도와 단점
##### - 시간복잡도(탐색 시)
- depth를 h라 하면, O(h)
- N개의 node를 가진다면, h = logN 에 가까우므로, O(logN)
