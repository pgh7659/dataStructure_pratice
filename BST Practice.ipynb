```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class NodeMgmt:
    def __init__(self, head):
        self.head = head
        
    def insert(self, value):
        self.current_node = self.head
        self.parent_node = self.head
        
        while True:
            if self.current_node.value > value:
                self.parent_node = self.current_node
                self.current_node = self.current_node.left
            else:
                self.parent_node = self.current_node
                self.current_node = self.current_node.right
                
            if self.current_node == None:
                break
        
        self.current_node = Node(value)
        if self.parent_node.value > value:
            self.parent_node.left = self.current_node
        else:
            self.parent_node.right = self.current_node
        
    def search(self, value):
        self.current_node = self.head
        
        while self.current_node:
            if self.current_node.value == value:
                return self.current_node
            elif self.current_node.value > value:
                self.current_node = self.current_node.left
            else:
                self.current_node = self.current_node.right
        return None

    def delete(self, value):
        self.current_node = self.head
        self.parent_node = self.head
        self.deletable = False
        
        while self.current_node:
            if self.current_node.value == value:
                self.deletable = True
                break
            elif self.current_node.value > value:
                self.parent_node = self.current_node
                self.current_node = self.current_node.left
            else:
                self.parent_node = self.current_node
                self.current_node = self.current_node.right
        
        if self.deletalbe:
            # child node가 없는 경우
            if self.current_node.left == None and self.current_node.right == None:
                # current_node가 parent_node의 left child인 경우
                if self.parent_node.value > value:
                    self.parent_node.left = None
                # current_node가 parent_node의 right child인 경우
                else:
                    self.parent_node.right = None
                
                del self.current_node
            
            # child node가 1개(left child node)인 경우
            elif self.current_node.left != None or self.current_node.right == None:
                if self.parent_node.value > value:
                    self.parent_node.left = self.current_node.left
                else:
                    self.parent_node.right = self.current_node.left
                    
                del self.current_node
                
            # child node가 1개(right child node)인 경우
            elif self.current_node.left == None or self.current_node.right != None:
                if self.parent_node.value > value:
                    self.parent_node.left = self.current_node.right
                else:
                    self.parent_node.right = self.current_node.right
                    
                del self.current_node

            # child node가 2개인 경우
            else:
                # 큰 값중 최소값 적용
                
                # 삭제노드의 대체노드 초기값 설정: "큰값중 최소값"이므로 삭제노드의 오른쪽(큰값쪽)이 초기값
                change_node = self.current_node.right
                change_parent = self.current_node
                
                # 삭제노드의 대체노드 검색: "큰값중 최소값"이므로 삭제노드의 오른쪽 자식노드부터 왼쪽노드만 검색
                while change_node.left:
                    change_parent = change_node
                    change_node = change_node.left
                
                # 대체노드의 오른쪽 자식노드 여부 확인
                # 있는 경우: 대체노드의 부모노드 왼쪽 자식노드로 이어준다
                if change_node.right:
                    change_parent.left = change_node.right
                else:
                    change_parent.left = None
                
                # 대체노드의 자식노드들을 삭제될 노드의 자식 노드들로 설정
                change_node.left = self.current_node.left
                change_node.right = self.current_node.right
                
                # 삭제노드의 부모노드에 대체 노드 연결
                # 삭제노드의 위치가 부모노드의 왼쪽인지 오른쪽인지 결정
                if self.parent_node.value > value:
                    self.parent_node.left = change_node
                else:
                    self.parent_node.right = change_node
```

```python
head = Node(50)
BST = NodeMgmt(head)

BST.insert(30)
BST.insert(75)
BST.insert(15)
BST.insert(100)
```
